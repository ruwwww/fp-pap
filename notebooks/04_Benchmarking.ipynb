# ⚡ Notebook 04: Benchmarking (All Models at Once)

**Goal**: Run ALL enabled models in `models_config.yaml` across all 3 scenarios in one command.

This is the scalable 'fire and forget' approach for quick comparison.

```python
import sys
sys.path.insert(0, '..')

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from benchmarking import BenchmarkRunner

print('✅ BenchmarkRunner ready')
```

```python
# ── Load Data ─────────────────────────────────────────────────────────
df = pd.read_csv('data/raw/train.csv')  # ← UPDATE
df['date'] = pd.to_datetime(df['date']) # ← UPDATE date column name
df = df.sort_values('date').reset_index(drop=True)
print(f'Dataset: {df.shape}')
```

```python
# ── Run Full Benchmark ────────────────────────────────────────────────
# Tip: To skip slow models: skip_models=['lstm', 'gru']
# Tip: To run only fast ML: only_models=['xgboost', 'lightgbm']

runner = BenchmarkRunner(
    config_path='config/config.yaml',
    models_config_path='config/models_config.yaml',
    use_mlflow=True,
)

results = runner.run(
    df,
    # skip_models=['svr'],        # optional: skip slow models
    # only_models=['xgboost'],    # optional: quick smoke test
)
```

```python
# ── Save all results ──────────────────────────────────────────────────
runner.save_results()

# ── Show best models ──────────────────────────────────────────────────
print('\n🏆 Best per scenario:')
runner.best_per_scenario()
```

```python
# ── Full results table ────────────────────────────────────────────────
df_results = results.to_dataframe()
df_results.style.background_gradient(subset=['RMSE', 'MAPE (%)'], cmap='RdYlGn_r')
```
